Q3.7 comment:

Let me briefly explain the questions. One might expect that a stronger Pearson correlation would be related to a smaller Euclidean distance. If two objects are similar, they might appear in closer proximity with a smaller Euclidean distance. While these are reasonable intuition, the Euclidean distance is a absolute distance metric whereas Pearson correlation and cosine are normalized to [-1, 1].

Theoretically, both Pearson corr and cosine are computed by covariance divided by the product of deviations. By default, however, cosine measures these from the origin where Pearson corr is from the means. When cosine is based on the departure from the mean, as in question 3, it is equivalent to Pearson corr.
