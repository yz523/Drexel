```python
def is_prime(n):
   if (n < 2):
     return False
   d = 2
   while d*d <= n:
     if (n % d) == 0:
       return False
     d = d+1
   return True

def sum_primes(a, b):
  accum = 0
  for x in range(a, b+1):
    if is_prime(x):
      accum += x
  return accum

def my_range(a, b):
  i = a
  R = []
  while i < b:
    R.append(i)
    i = i + 1
  return R

class MyNumbers:
  def __init__(self, start, end):
    self.a0 = start
    self.b = end

  def __iter__(self):
    self.a = self.a0
    return self

  def __next__(self):
    if self.a < self.b:
      x = self.a
      self.a += 1
      return x
    else:
      raise StopIteration

def my_range_delayed(a, b):
  i = a
  while i < b:
    yield i
    i = i + 1
```

###Exercise 5(Part 1)

```python
class MyMapIter:
  def __init__(self, f, L):
    self.index = 0
    self.L = L
    self.f = f

  def __iter__(self):
    self.index = 0
    return self

  def __next__(self):
    if self.index < len(self.L):
      x = self.f(self.L[self.index])
      self.index += 1
      return x
    else:
      raise StopIteration

print(list(map(square,range(1,10))))
print(list(MyMapIter(square,range(1,10))))
for i in MyMapIter(square,range(1,10)):
  print(i)
```

```python
%%file maxword.py

from mrjob.job import MRJob
from mrjob.step import MRStep
import re

WORD_RE = re.compile(r"[\w']+")


class MRMostUsedWord(MRJob):

    def mapper_get_words(self, _, line):
        # yield each word in the line
        for word in WORD_RE.findall(line):
            yield (word.lower(), 1)

    def combiner_count_words(self, word, counts):
        # sum the words we've seen so far
        yield (word, sum(counts))

    def reducer_count_words(self, word, counts):
        # send all (num_occurrences, word) pairs to the same reducer.
        # num_occurrences is so we can easily use Python's max() function.
        yield None, (sum(counts), word)

    # discard the key; it is just None
    def reducer_find_max_word(self, _, word_count_pairs):
        # each item of word_count_pairs is (count, word),
        # so yielding one results in key=counts, value=word
        yield max(word_count_pairs)

    def steps(self):
        return [
            MRStep(mapper=self.mapper_get_words,
                   combiner=self.combiner_count_words,
                   reducer=self.reducer_count_words),
            MRStep(reducer=self.reducer_find_max_word)
        ]


if __name__ == '__main__':
    MRMostUsedWord.run()
```

```python
%%file wordsSort.py

from mrjob.job import MRJob
from mrjob.step import MRStep
import re

WORD_RE = re.compile(r"[\w']+")


class MRWordFreqSort(MRJob):

    def mapper_get_words(self, _, line):
        
        for word in WORD_RE.findall(line):
            yield (word.lower(), 1)

    def combiner_count_words(self, word, counts):
        yield (word, sum(counts))

    def reducer_count_words(self, word, counts):
        yield None, (sum(counts), word)

    def reducer_sort_word(self, _, word_count_pairs):
        for count, word in sorted(word_count_pairs, reverse=True):
            yield (word, count)

    def steps(self):
        return [
            MRStep(mapper=self.mapper_get_words,
                   combiner=self.combiner_count_words,
                   reducer=self.reducer_count_words),
            MRStep(reducer=self.reducer_sort_word)
        ]


if __name__ == '__main__':
    MRWordFreqSort.run()
```

```python
%%file wordsKMax.py

from mrjob.job import MRJob
from mrjob.step import MRStep
import re
import sys

WORD_RE = re.compile(r"[\w']+")



class MRMostKUsedWord(MRJob):
    
    def configure_args(self):
        super(MRMostKUsedWord, self).configure_args()
        self.add_passthru_arg('--k', type = int, default = 1)
    
    def load_args(self, args):
        super(MRMostKUsedWord, self).load_args(args)
    
    def mapper_get_words(self, _, line):
        for word in WORD_RE.findall(line):
            yield (word.lower(), 1)

    def combiner_count_words(self, word, counts):
        yield (word, sum(counts))

    def reducer_count_words(self, word, counts):
        yield None, (sum(counts), word)

    def reducer_sort_word(self, _, word_count_pairs):
        index = 0
        for count, word in sorted(word_count_pairs, reverse=True):
                index = index + 1
                if index <= int(self.options.k):
                    yield (word, count)

    def steps(self):
        return [
            MRStep(mapper=self.mapper_get_words,
                   combiner=self.combiner_count_words,
                   reducer=self.reducer_count_words),
            MRStep(reducer=self.reducer_sort_word)
        ]


if __name__ == '__main__':
    MRMostKUsedWord.run()
```
